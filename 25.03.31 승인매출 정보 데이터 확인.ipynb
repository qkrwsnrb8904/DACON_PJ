- `납부_보험료이용금액`: 최근 3개월 동안 보험료를 납부한 총 금액.
- `납부_건강연금이용금액`: 최근 3개월 동안 건강보험 및 연금을 납부한 총 금액.
- `RP건수_보험_B0M`: 이번 달 보험료 정기결제(RP) 건수.
- `RP건수_건강_B0M`: 이번 달 건강 관련 정기결제(RP) 건수.
- `RP후경과월_보험`: 마지막 보험 정기결제 이후 경과한 개월 수.
- `RP후경과월_건강`: 마지막 건강 정기결제 이후 경과한 개월 수.
- `증감_RP건수_보험_전월`: 지난달 대비 보험 정기결제 건수 증감량.
- `증감_RP건수_건강_전월`: 지난달 대비 건강 정기결제 건수 증감량.
- `정상청구원금_B0M`: 이번 달 정상적으로 청구된 원금 금액.
- `선입금원금_B0M`: 이번 달 선입금된 원금 금액.
- `정상입금원금_B0M`: 이번 달 정상적으로 입금된 원금 금액.
